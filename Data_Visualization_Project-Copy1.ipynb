```python
sea = sns.FacetGrid(df, col= "chronic Lung Disease", height = 4)
sea.map(sns.distplot, "Genetic Risk")
```

```python
sea = sns.FacetGrid(df, col= "chronic Lung Disease", height = 4)
sea.map(sns.histplot, "Air Pollution")
```

```python
sea = sns.FacetGrid(df, col= "chronic Lung Disease", height = 4)
sea.map_dataframe(sns.barplot, x="Air Pollution", y="Age")
```
